### Skewness Example Based On Multiplication Factor

```python
# import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load titanic dataset
titanic_url = r"https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df  = pd.read_csv(titanic_url)

#normilize the column using min-max scaler
def normelize(Column):
  return (Column - Column.min())/ (Column.max() - Column.min())

df['normelized_age'] = normelize(df['Age'])

#creaet a left skewed cloumn with outliers
df['left_skewed'] = df['normelized_age']*0.5 # Multiply by factor less than 1

#creaet a left skewed cloumn with outliers
df['right_skewed'] = df['normelized_age']*2 # Multiply by factor greater than 1

#plot
plt.figure(figsize=(10,6))
sns.boxplot(data = df[['normelized_age','left_skewed','right_skewed']])
plt.ylim(0,1.3) # set y-axis range
plt.xlabel('Cloumn')
plt.ylabel('Value')
plt.title('Box Or Whisker Plot')
plt.show()
```

### Skewness Example Based On Addition and Subtraction Factor

```python
# import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load titanic dataset
titanic_url = r"https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df  = pd.read_csv(titanic_url)

#normilize the column using min-max scaler
def normelize(Column):
  return (Column - Column.min())/ (Column.max() - Column.min())

df['normelized_age'] = normelize(df['Age'])

#creaet a left skewed cloumn with outliers
df['left_skewed'] = df['normelized_age'] - 0.2  #shifting left skewed
df.loc[df['left_skewed'] < -0.2 , 'left_skewed'] -= 0.5  # Add outliers

#creaet a left skewed cloumn with outliers
df['right_skewed'] = df['normelized_age'] + 0.2  #shifting right skewed
df.loc[df['right_skewed'] > 1.2 , 'right_skewed'] += 0.5  # Add outliers

#plot
plt.figure(figsize=(10,7))
sns.boxplot(data = df[['normelized_age','left_skewed','right_skewed']])
plt.ylim(0,1.2) # set y-axis range
plt.xlabel('Cloumn')
plt.ylabel('Value')
plt.title('Box Or Whisker Plot')
plt.show()
```

### Skewed Example 

```python
#import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load the iris dataset
iris_url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(iris_url)

#normelize a column using min-max scaling
def normelize(Column):
 return (Column - Column.min())/ (Column.max() - Column.min())

df['normelized_sepal_witdh'] = normelize(df['sepal_width'])

#Left_Skwed_data
left_skew_data = np.random.normal(loc = -0.25 , scale = 0.05, size = len(df))
df['left_skewed'] = df['normelized_sepal_witdh'] - left_skew_data

#Right_Skwed_data
right_skew_data = np.random.normal(loc = 0.2 , scale = 0.05, size = len(df))
df['right_skewed'] = df['normelized_sepal_witdh'] - right_skew_data

#plot
plt.figure(figsize=(10,7))
sns.boxplot(data = df[['normelized_sepal_witdh','left_skewed','right_skewed']])
plt.ylim(-0.2,1.2) # set y-axis range
plt.xlabel('Cloumn')
plt.ylabel('Value')
plt.title('Box Or Whisker Plot')
plt.show()
```
